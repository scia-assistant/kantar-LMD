## Clustering choice

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
import matplotlib.pyplot as plt


def calculate_variance_ratios(data, labels):
    overall_mean = data.mean(axis=0)
    groups = np.unique(labels)
    
    intra_variance = 0
    inter_variance = 0
    
    for group in groups:
        group_data = data[labels == group]
        group_mean = group_data.mean(axis=0)
        
        intra_variance += np.sum((group_data - group_mean) ** 2)
        
        inter_variance += len(group_data) * np.sum((group_mean - overall_mean) ** 2)

    return intra_variance, inter_variance, intra_variance / inter_variance

def clustering_compare(data):
    
    inertia = []
    silhouette_scores = []
    k_values = range(2, 11)
    variance_ratios = []

    for k in k_values:
        kmeans = KMeans(n_clusters=k, random_state=42)
        labels = kmeans.fit_predict(data)
    
        inertia.append(kmeans.inertia_)
        silhouette_scores.append(silhouette_score(data, labels))
        intra, inter, ratio = calculate_variance_ratios(data.values, labels)
        variance_ratios.append(ratio)
        print(f'k={k}: Intra-variance={intra}, Inter-variance={inter}, Ratio={ratio:.4f}')

    plt.figure(figsize=(8, 5))
    plt.plot(k_values, inertia, marker='o')
    plt.title('Méthode du coude')
    plt.xlabel('Nombre de clusters (k)')
    plt.ylabel('Inertie intra-groupe')
    plt.show()

    plt.figure(figsize=(8, 5))
    plt.plot(k_values, silhouette_scores, marker='o')
    plt.title('Indice de silhouette')
    plt.xlabel('Nombre de clusters (k)')
    plt.ylabel('Score de silhouette')
    plt.show()
    
    plt.figure(figsize=(8, 5))
    plt.plot(k_values, variance_ratios, marker='o')
    plt.title('Ratio variances intra/inter groupes')
    plt.xlabel('Nombre de clusters (k)')
    plt.ylabel('Ratio (intra/inter)')
    plt.show()
    
    linked = linkage(data, method='ward')
    
    plt.figure(figsize=(10, 7))
    dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=False)
    plt.title('Dendrogramme (Clustering hiérarchique)')
    plt.xlabel('Individus')
    plt.ylabel('Distance')
    plt.show()
    
    optimal_k_hierarchical = 3
    hierarchical_labels = fcluster(linked, t=optimal_k_hierarchical, criterion='maxclust')
    
    intra, inter, ratio = calculate_variance_ratios(data.values, hierarchical_labels)
    print(f'Hiérarchique: Intra-variance={intra}, Inter-variance={inter}, Ratio={ratio:.4f}')
```

### Orange

```python
clustering_compare(orange_normalized)
```

#### Obsevation

Le ratio intra/inter-variances diminue avec k, ce qui est attendu, mais la réduction est moins significative après k=5.
k=5 semble offrir un bon compromis, avec un ratio 0.9637.

Clustering Hiérarchique (k=3) :
Le ratio pour k=3 est 1.1283, légèrement supérieur à celui de K-Means (1.1275).
Cela indique que K-Means offre une meilleure séparation des groupes pour ce nombre de clusters.
Conclusion :

K-Means avec k=5 est probablement le meilleur choix pour une segmentation plus fine.
Si une solution plus simple est souhaitée, K-Means avec k=3 pourrait être une alternative compétitive.

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

data = orange_normalized

k = 5
kmeans = KMeans(n_clusters=k, random_state=42)
labels = kmeans.fit_predict(data)

intra, inter, ratio = calculate_variance_ratios(data.values, labels)
print(f'K-Means (k={k}): Intra-variance={intra:.4f}, Inter-variance={inter:.4f}, Ratio={ratio:.4f}')

data['Cluster'] = labels

pca = PCA(n_components=2)
data_pca = pca.fit_transform(data.drop(columns=['Cluster']))

plt.figure(figsize=(8, 6))
plt.scatter(data_pca[:, 0], data_pca[:, 1], c=labels, cmap='viridis', s=50, alpha=0.7)
plt.title('Visualisation des clusters K-Means (k=5) avec PCA')
plt.xlabel('Composante principale 1')
plt.ylabel('Composante principale 2')
plt.colorbar(label='Cluster')
plt.show()
```

Hiérarchique avec k=3 :
  - Ratio : 2.3512, légèrement supérieur à celui obtenu avec K-Means (2.2144).

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

data = green_normalized

k = 6
kmeans = KMeans(n_clusters=k, random_state=42)
labels = kmeans.fit_predict(data)

intra, inter, ratio = calculate_variance_ratios(data.values, labels)
print(f'K-Means (k={k}): Intra-variance={intra:.4f}, Inter-variance={inter:.4f}, Ratio={ratio:.4f}')

data['Cluster'] = labels

pca = PCA(n_components=2)
data_pca = pca.fit_transform(data.drop(columns=['Cluster']))

plt.figure(figsize=(8, 6))
plt.scatter(data_pca[:, 0], data_pca[:, 1], c=labels, cmap='viridis', s=50, alpha=0.7)
plt.title('Visualisation des clusters K-Means (k=5) avec PCA')
plt.xlabel('Composante principale 1')
plt.ylabel('Composante principale 2')
plt.colorbar(label='Cluster')
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

X = data.drop(columns=['Cluster'])  # Variables actives
y = data['Cluster']  # Labels des clusters

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

feature_importance = pd.Series(model.feature_importances_, index=X.columns).sort_values(ascending=False)
print("Importance des variables :")
print(feature_importance)

y_pred = model.predict(X_test)

print("Rapport de classification :")
print(classification_report(y_test, y_pred))

print("Matrice de confusion :")
print(confusion_matrix(y_test, y_pred))
```
